```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")

text = "The course content is amazing and very useful."
result = sentiment(text)

print(result)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)

text = "Rishik Vihaan studies Computer Science at RV University in Bangalore."
entities = ner(text)

for e in entities:
    print(e)
```

```python
translator = pipeline("translation_en_to_fr")

text = "Machine learning is transforming the world."
translated = translator(text)

print(translated[0]['translation_text'])
```

```python
summarizer = pipeline("summarization")

text = """
Artificial Intelligence is a branch of computer science that aims to create
machines capable of performing tasks that normally require human intelligence.
"""

summary = summarizer(text, max_length=50, min_length=20)
print(summary[0]['summary_text'])
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "I love machine learning",
    "Deep learning is fascinating",
    "I enjoy playing football"
]

embeddings = model.encode(sentences)

similarity = util.cos_sim(embeddings[0], embeddings[1])
print(similarity)
```

```python
pip install fastai
```

```python
from fastai.vision.all import *
import fastprogress.fastprogress as fp

# Completely disable progress bars
fp.NO_BAR = True

path = untar_data(URLs.PETS)/'images'

def is_cat(x):
    return x[0].isupper()

dls = ImageDataLoaders.from_name_func(
    path,
    get_image_files(path),
    valid_pct=0.2,
    label_func=is_cat,
    item_tfms=Resize(224)
)

learn = vision_learner(dls, resnet18, metrics=accuracy)

# 🔑 FIX: Remove problematic progress callback
learn.remove_cb(ProgressCallback)

learn.fine_tune(1)

learn.show_results()
```

3. Hugging Face AI – Simple Practical Application

```python
from transformers import pipeline

classifier = pipeline("text-classification")

while True:
    text = input("Enter a sentence (or 'exit'): ")
    if text.lower() == 'exit':
        break
    result = classifier(text)
    print(result)
```

```python
qa = pipeline("question-answering")

context = """
FastAI is a deep learning library that simplifies training neural networks.
"""

question = "What is FastAI?"

answer = qa(question=question, context=context)
print(answer['answer'])
```
