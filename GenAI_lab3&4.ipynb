```python
import os
from getpass import getpass

os.environ["GROQ_API_KEY"] = getpass("Enter Groq API Key: ")
```

```python
from groq import Groq
import os

client = Groq(
    api_key=os.environ["GROQ_API_KEY"]
)
```

```python
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "Say hello"}]
)

print(response.choices[0].message.content)
```

```python
def ask_ai(title, prompt):
    print(f"\n===== {title} =====\n")

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    print(response.choices[0].message.content)
```

```python
ask_ai(
    "System Check",
    "Explain in one sentence what Prompt Engineering is."
)
```

1. Persona Prompt

```python
ask_ai(
    "Persona Prompt",
    """
    You are a cybersecurity expert with 15 years of experience.
    Explain phishing attacks to a beginner.
    """
)
```

2. Cognitive Verifier Pattern

```python
ask_ai(
    "Cognitive Verifier Pattern",
    """
    Solve the problem step-by-step and verify each step.

    Find the sum of first 10 natural numbers.
    """
)
```

3. Question Refinement Pattern

```python
ask_ai(
    "Question Refinement Pattern",
    """
    The user asks: 'Tell me about Artificial Intelligence.'

    Refine this into 3 clearer questions before answering.
    """
)
```

4. Provide New Information and Ask Questions

```python
ask_ai(
    "Provide Info + Ask Questions",
    """
    Explain Machine Learning briefly.
    After explanation, ask 3 questions to test understanding.
    """
)
```

5. Root Prompt

```python
ask_ai(
    "Root Prompt",
    """
    You are an AI tutor.
    Explain DBMS step-by-step.
    Use simple language.
    Provide real-life examples.
    Avoid technical jargon.
    """
)
```

6. Chain of Thought Prompting

```python
ask_ai(
    "Chain of Thought",
    """
    Solve step-by-step:

    A train travels 60 km in 1 hour.
    How far will it travel in 5 hours?
    """
)
```

7. Tabular Format Prompting

```python
ask_ai(
    "Tabular Format",
    """
    Create a table comparing:
    Artificial Intelligence,
    Machine Learning,
    Deep Learning

    Based on:
    Definition
    Complexity
    Applications
    """
)
```

8. Fill in the Blank Prompt

```python
ask_ai(
    "Fill in the Blank",
    """
    Fill in the blank:

    HTML stands for _________
    """
)
```

9. RGC Framework Prompt

```python
ask_ai(
    "RGC Framework",
    """
    Role: Data Scientist
    Goal: Explain Linear Regression
    Constraints: Use simple language and real-life examples
    """
)
```

10. Zero Shot Prompting

```python
ask_ai(
    "Zero Shot Prompting",
    """
    Translate 'Good Morning' into French.
    """
)
```

11. One Shot Prompting

```python
ask_ai(
    "One Shot Prompting",
    """
    Example:
    Dog -> Animal

    Now classify:
    Rose ->
    """
)
```

12. Few Shot Prompting

```python
ask_ai(
    "Few Shot Prompting",
    """
    Examples:
    Cat -> Animal
    Rose -> Plant
    Car -> Vehicle

    Now classify:
    Mango ->
    """
)
```
